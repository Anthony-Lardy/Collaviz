```python
import re
import pandas as pd
import csv
from datetime import datetime

def is_create(line):
    return 'CREATE TABLE' in line or False

def is_insert(line):
    return 'INSERT INTO' in line or False

def is_finish(line):
    return ';' in line or False

def get_table_name_create(line):
    match = re.search('CREATE TABLE IF NOT EXISTS `([0-9_a-zA-Z]+)`', line)
    if match:
        return match.group(1)
    else:
        print(line)
        
def get_table_name_insert(line):
    match = re.search('INSERT INTO `([0-9_a-zA-Z]+)`', line)
    if match:
        return match.group(1)
    else:
        print(line)

def create_csv_file(filename):
    filename += ".csv"
    open(filename, 'w')

def set_values_insert(line, file):
    #s = ","
    p = re.compile('`([0-9_a-zA-Z]+)`')
    writer = csv.writer(file)
    writer.writerow(p.findall(line)[1:])
                    
def sql_to_csv(sqlfile):
    reader = open(sqlfile, 'r', encoding='utf-8')
    contenu = reader.read()
    lines = contenu.split("\n")
    nb_table = 0
    insert = False
    for line in lines:
        if insert and (not is_insert(line)):
            writer = csv.writer(file)
            writer.writerow(line[1:-2].split(", "))
        if is_create(line):
            create_csv_file(get_table_name_create(line))
            nb_table += 1
        if is_insert(line):
            filename = get_table_name_insert(line)
            insert = True
            file = open(filename+".csv", 'a')
            set_values_insert(line, file) 
        if is_finish(line):
            insert = False
    print("Il y a", nb_table, "tables")
        
deb = datetime.now()
sql_to_csv("collaviz/traceforum.sql")
fin = datetime.now()
print(deb, fin)
```

```python
data = pd.read_csv("transition.csv", encoding='utf-8')
def getReponsesPost(data, idParent):
    utilisateur = data['Utilisateur'][(data['Attribut'].str.contains('IDParent='+idParent)) & (data['Titre'] == "'Répondre à un message'")]
    attribut = data['Attribut'][(data['Attribut'].str.contains('IDParent='+idParent))]
    if(attribut.empty):
        return ("", "")
    idMessage = str(attribut).split(",")[1].split("\n")[0].split("=")[1]
    return (utilisateur, idMessage)
print(getReponsesPost(data, '1163'))
```

```python
#pour chaque utilisateur
import numpy as np
deb = datetime.now()
utilisateurs = data["Utilisateur"].unique().tolist()
for l in range(len(utilisateurs)):
    print("\n Pour l'utilisateur :" + utilisateurs[l] +"\n" )
    #on initialise la matrice de résultat
    a = np.zeros(shape=(1,len(utilisateurs)))
    tableauResultat = pd.DataFrame(a, index = ['l1'], columns = utilisateurs)
    #on recupere les posts qu'il a créé
    posts = postsUtilisateur(data, utilisateurs[l])
    
    #pour chaque post qu'il a créé
    for i in range(len(posts)):
        tableauUsers = []
        user, idMessage = getReponsesPost(data, posts[i])
        #tant qu'il y a des messages on déroule le fil de la discussion
        while idMessage != "" :  
            #on récupere l'identifiant du message et l'utilisateur qui a posté le message
            tableauUsers.append(user.values[0])
            user, idMessage = getReponsesPost(data, idMessage)
        tableauUsers = list(set(tableauUsers))
       
        # on ajoute 1 dans la matrice si l'utilisateur a répondu au post
        for j in range(len(utilisateurs)):
            for k in range(len(tableauUsers)):
                if tableauUsers[k] == utilisateurs[j]:
                    tableauResultat.at['l1', utilisateurs[j]] = (1 + tableauResultat.at['l1', utilisateurs[j]])
   
    # on calcule la matrice en pourcentage
    if (len(posts) != 0) :
        tableauResultat = (tableauResultat * 100) / len(posts)
    print(tableauResultat)
fin = datetime.now()
print(deb)
print(fin)
```

```python
data = pd.read_csv("transition.csv", encoding='utf-8')
def getReponses(data, action, utilisateur):
     return(data[(data['Titre'] == action) & (data['Utilisateur'] == utilisateur)])

def getMessage(data, idMsg):
     return(data[((data['Titre'].str.contains('Répondre à un message'))
                              |data['Titre'].str.contains('Poster un nouveau message'))
                             & data['Attribut'].str.contains('IDMsg='+idMsg)])

from datetime import datetime, date

def getDifferenceMessages(dateMessage, timeMessage, dateParent, timeParent):
    return int((datetime.combine(datetime(dateMessage).date, timeMessage) - datetime.combine(datetime(dateParent).date, timeParent)).total_seconds())


def temps_moyen(seconds):
    intervals = (
    ('semaines', 604800),  # 60 * 60 * 24 * 7
    ('jours', 86400),    # 60 * 60 * 24
    ('heures', 3600),    # 60 * 60
    ('minutes', 60),
    ('secondes', 1),
    )
    result = []
    for name, count in intervals:
        value = seconds // count
        if value:
            seconds -= value * count
            if value == 1:
                name = name.rstrip('s')
            result.append("{} {}".format(value, name))
    return ', '.join(result[:5])
```

```python
import numpy as np
dataFrame = pd.read_csv("transition.csv", encoding='utf-8')

# A remplacer Titre par Action  
def actionDate(dataFrame, Action, Utilisateur, Date):
    return dataFrame[(dataFrame['Titre'] == Action) & (dataFrame['Utilisateur'] == Utilisateur) & (dataFrame['Date'] == Date)].count()['Date']
   
def action2Date(dataFrame, Action, Utilisateur, Date1,Date2):
    List=[]
    
    for d in dataFrame['Date'].unique().tolist():
        if ((d >= Date1) and (d <= Date2)):
            List.append(actionDate(dataFrame, Action, Utilisateur, d))
    return sum(List)


def actionParTemps(dataFrame, Action, Utilsateurs, Date1, Date2):
    
    List=[]
    list_nbAction=[]
    if Date1 is None:
        for user in Utilsateurs:
            for d in dataFrame['Date'].unique().tolist():
                if (d <= Date2):  
                    list_nbAction.append(actionDate(dataFrame, Action, user, d))
            List.append((user,list_nbAction))
    elif Date2 is None:
        for user in Utilsateurs:
            for d in dataFrame['Date'].unique().tolist():
                if (d >= Date1):  
                    list_nbAction.append(actionDate(dataFrame, Action, user, d))
            List.append((user,list_nbAction))
    elif (Date2 is None) and (Date1 is None):
        for user in Utilsateurs:
            for d in dataFrame['Date'].unique().tolist():             
                list_nbAction.append(actionDate(dataFrame, Action, user, d))
            List.append((user,list_nbAction))
    else:
        for user in Utilsateurs:
            for d in dataFrame['Date'].unique().tolist():
                if ((d >= Date1) and (d <= Date2)):  
                    list_nbAction.append(actionDate(dataFrame, Action, user, d))
            List.append((user,list_nbAction))

    return List

List=actionParTemps(dataFrame, "'Répondre à un message'",["'tdelille'","'mmay'"], "'2009-02-12'", "'2009-02-13'")
print(List)
```

```python
from collections import OrderedDict, defaultdict

def duppliCellulelist(fichier, colonne, separateur):
    data = pd.read_csv(fichier, encoding = "latin-1")
    row_list = []
    for index, row in data.iterrows():
        if len(row[colonne].split(separateur)) > 1:
            for i in row[colonne].split(separateur):
                row[colonne] = i
                row_list.append(row.to_dict(OrderedDict))
        else:
            row_list.append(row.to_dict(OrderedDict))
    newData = pd.DataFrame(row_list, columns=data.columns)
    return newData

dataFrame=duppliCellulelist("transition.csv", "Attribut", ',')
dataFrame.head()
```
