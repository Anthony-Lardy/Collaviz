```python
import re
import pandas as pd
import csv
from datetime import datetime

def is_create(line):
    return 'CREATE TABLE' in line or False

def is_insert(line):
    return 'INSERT INTO' in line or False

def is_finish(line):
    return ';' in line or False

def get_table_name_create(line):
    match = re.search('CREATE TABLE IF NOT EXISTS `([0-9_a-zA-Z]+)`', line)
    if match:
        return match.group(1)
    else:
        print(line)
        
def get_table_name_insert(line):
    match = re.search('INSERT INTO `([0-9_a-zA-Z]+)`', line)
    if match:
        return match.group(1)
    else:
        print(line)

def create_csv_file(filename):
    filename += ".csv"
    open(filename, 'w')

def set_values_insert(line, file):
    #s = ","
    p = re.compile('`([0-9_a-zA-Z]+)`')
    writer = csv.writer(file)
    writer.writerow(p.findall(line)[1:])
                    
def sql_to_csv(sqlfile):
    reader = open(sqlfile, 'r', encoding='utf-8')
    contenu = reader.read()
    lines = contenu.split("\n")
    nb_table = 0
    insert = False
    for line in lines:
        if insert and (not is_insert(line)):
            writer = csv.writer(file)
            writer.writerow(line[1:-2].split(", "))
        if is_create(line):
            create_csv_file(get_table_name_create(line))
            nb_table += 1
        if is_insert(line):
            filename = get_table_name_insert(line)
            insert = True
            file = open(filename+".csv", 'a')
            set_values_insert(line, file) 
        if is_finish(line):
            insert = False
    print("Il y a", nb_table, "tables")
        
deb = datetime.now()
sql_to_csv("traceforum.sql")
fin = datetime.now()
print(deb, fin)
```

```python
data = pd.read_csv("transition.csv", encoding='latin-1') 
def actionDate(data, Action, Utilisateur, Date):
    return data[(data['Titre'] == Action) & (data['Utilisateur'] == Utilisateur) & (data['Date'] == datetime.strptime(Date, "'%Y-%m-%d'").date())].count()['Date']

print(actionDate(data, "'Connexion'", "'tdelille'","'2009-02-12'"))
print(actionDate(data, "'Afficher le contenu d''un message'", "'mmay'","'2009-02-13'"))
print(actionDate(data, "'Répondre à un message'", "'mmay'","'2009-02-13'"))
```
